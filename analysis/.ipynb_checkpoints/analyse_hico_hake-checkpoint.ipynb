```python
import os
os.chdir('/home/alex/PartHOIZS')

from matplotlib import pyplot as plt
import numpy as np

from analysis.utils import plot_mat
from lib.dataset.hicodet_hake import HicoDetHake
from lib.dataset.word_embeddings import WordEmbeddings
```

```python
hh = HicoHake()
split = 'train'
# split = 'test'

inters = hh.split_annotations[split]
actions = np.minimum(1, inters @ hh.interaction_to_action_mat)
fg_part_states = np.concatenate([inds[:-1] for inds in hh.states_per_part])
```

```python
part_states = hh.split_part_annotations[split]

use_pearson = False
if not use_pearson:  # co-occurrences:
    pstate_inters_cooccs = part_states[:, fg_part_states].T @ inters
    pstate_inters_cooccs /= pstate_inters_cooccs.sum(axis=0, keepdims=True)
    m = pstate_inters_cooccs
    part_state_strs = [f'{p} {a}'
                        for p, a in hh.bp_ps_pairs 
                       if a != hh.null_action
                       ]
else:  # Pearson's correlation
    m = np.array([[np.corrcoef(x, y)[1, 0] for y in inters.T] 
                  for x in part_states.T])
    part_state_strs = [f'{p} {a}'
                        for p, a in hh.bp_ps_pairs 
                       # if a != hh.null_action
                       ]

inter_strs = [f'{hh.actions[a] if a > 0 else "NULL"} {hh.objects[o]}' 
              for a, o in hh.interactions]

for i in range(6):
    plot_mat(m[:, i*100:(i+1)*100], 
             xticklabels=inter_strs[i*100:(i+1)*100], 
             yticklabels=part_state_strs,
             zero_color=[1, 1, 1],
             vrange=(-1, 1) if use_pearson else (0, 1))
    
```

```python
bp = hh.part_inv_ind['head']
for i in hh.states_per_part[bp]:
    if hh.bp_ps_pairs[i][1] == 'blow':
        ps = i
        break
else:
    raise ValueError
i = np.flatnonzero(np.all(hh.interactions == np.array([hh.action_index['blow'], 
                                                       hh.object_index['cake']]),
                          axis=1))
print(m[ps, i])
    
```

```python
part_states = hh.split_part_annotations[split]

part_states_actions_cooccs = part_states.T @ actions
part_states_actions_cooccs /= part_states_actions_cooccs.sum(axis=0, keepdims=True)

part_state_strs = [f'{p} {a}'
                    for p, a in hh.part_state_pairs]

plot_mat(part_states_actions_cooccs, 
         xticklabels=hh.actions, 
         yticklabels=part_state_strs,
         zero_color=[1, 1, 1])
```

```python
part_actions = hh.split_part_annotations[split]
parts = np.stack([part_actions[:, inds[:-1]].any(axis=1) 
                  for inds in hh.states_per_part], 
                 axis=1)

part_actions_cooccs = parts.T @ actions
part_actions_cooccs /= part_actions_cooccs.sum(axis=0, keepdims=True)
# part_actions_cooccs /= np.sqrt(part_actions_cooccs.sum(axis=0, keepdims=True)) *\
#                        np.sqrt(part_actions_cooccs.sum(axis=1, keepdims=True)) 

plot_mat(part_actions_cooccs.T, 
         yticklabels=hh.actions, 
         xticklabels=hh.parts,
         zero_color=[1, 1, 1])
```

```python
part_actions = hh.split_part_annotations[split]
parts = np.stack([part_actions[:, inds[:-1]].any(axis=1) 
                  for inds in hh.states_per_part], 
                 axis=1)

part_interactions_cooccs = parts.T @ inters
part_interactions_cooccs /= part_interactions_cooccs.sum(axis=0, keepdims=True)
# part_actions_cooccs /= np.sqrt(part_actions_cooccs.sum(axis=0, keepdims=True)) *\
#                        np.sqrt(part_actions_cooccs.sum(axis=1, keepdims=True)) 

for i in range(6):
    plot_mat(part_interactions_cooccs[:, i*100:(i+1)*100], 
             xticklabels=hh.interactions_str[i*100:(i+1)*100], 
             yticklabels=hh.parts,
             zero_color=[1, 1, 1])
```

```python
word_embs = WordEmbeddings(source='glove', dim=300, normalize=True)

part_states_embs = word_embs.get_embeddings([a for p, a in hh.bp_ps_pairs], 
                                            retry='avg')
action_embs = word_embs.get_embeddings(hh.actions, retry='avg')

part_states_actions_emb_sim = part_states_embs @ action_embs.T

part_state_strs = [f'{p} {a}' for p, a in hh.bp_ps_pairs]
```

```python
plot_mat(part_states_actions_emb_sim, 
         xticklabels=hh.actions, 
         yticklabels=part_state_strs,
         zero_color=[1, 1, 1],
         bin_colours=True)
```

```python
null_interactions = (hh.interactions[:, 0] == 0)
i2o_mat = hh.interaction_to_object_mat[~null_interactions, :]
# objs = inters @ i2o_mat
objs = np.minimum(1, inters[:, ~null_interactions] @ i2o_mat)

num_objs_per_img = objs.sum(axis=1)
print(np.histogram(num_objs_per_img))

plt.hist(num_objs_per_img, bins=np.unique(num_objs_per_img), 
         rwidth=0.8, align='left', density=True)
plt.show()
```

```python
# Requires running a cell above
part_ranking_per_action = np.argsort(part_actions_cooccs.T, axis=1)[:, ::-1]

word_embs = WordEmbeddings(source='glove', dim=300, normalize=True)

part_embs = word_embs.get_embeddings([p for p in hh.parts], retry='avg')
action_embs = word_embs.get_embeddings(hh.actions, retry='avg')

part_action_sim = part_embs @ action_embs.T
emb_part_ranking_per_action = np.argsort(part_action_sim.T, axis=1)[:, ::-1]

top_ranks = np.stack([part_ranking_per_action[:, 0], 
                      emb_part_ranking_per_action[:, 0]], axis=0)
```
